### Choosing New Orleans zips

```python
df[df.CountyName == 'Orleans']
```

### Limiting to New Orleans zip codes

```python
dfnola = df[(df.CountyName == 'Orleans') & (df.State == 'LA')]
dfnola
```

```python
dfnola.iloc[2].isna().sum()
```

70122 has too many NaNs. Drop it.

```python
dfnola = dfnola.drop([1862])
dfnola.head()
```

Choosing zip in suburbs

```python
df[(df.CountyName == 'Jefferson') & (df.State =='LA')]
```

Add another zip code from the area. Suburb bordering New Orleans.

```python
dfmtr = df[(df.RegionName == 70005)]
dfmtr
```

```python
dfnola = dfnola.append(dfmtr)
dfnola.head()
```

### Re-format dataframe

```python
df =  dfnola.drop(columns=['RegionID', 'CountyName', 'City', 'State', 'Metro', 'SizeRank'])
df.head()
```

```python
df = df.set_index('RegionName')
df.index.name = None
df.head()
```

```python
df = df.transpose()
df.head()
```

Change index to datetime, and column names to strings

```python
df.columns = df.columns.astype(str)
df.index = pd.to_datetime(df.index, format='%Y-%m')
df
```

```python
df['70115'].iloc[1].dtype
```

```python
df.isna().sum()
```

No NaNs. Good.

## Check graphs and export dataframe

```python
font = {'family' : 'Verdana',
        'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)
plt.figure(figsize=(18,10))
plt.plot(df)
plt.legend(df.columns)
plt.show()
```

General trend is upwards except for 70131. where from 2004, prices are pretty stagnant. Top three zip codes start rising sharply around 2015.

```python
font = {'family' : 'Verdana',
        'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)
df.hist(figsize=(18,10))
plt.show()
```

These histograms show where the prices have leveled off for a long time (where they're highest) and where their spiking (where they're lowest). 

```python
# Reorder the columns so most expensive zips are first in the legend
df = df.reindex(columns=['70115', '70130', '70118', '70005', '70131'])
df.head()
```

```python
# Run line graph again
font = {'family' : 'Verdana',
        'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)
plt.figure(figsize=(18,10))
plt.plot(df)
plt.legend(df.columns)
plt.show()
```

```python
# Export csv
df.to_csv(r'../data/nola.csv', index=True)
```

```python
# Export line graph
font = {'family' : 'Verdana',
        'weight' : 'normal',
        'size'   : 12}
plt.rc('font', **font)
plt.figure(figsize=(18,10))
plt.plot(df)
plt.title('Mean Housing Prices - Five Area New Orleans Zip Codes - 1996-2018', fontsize=22, pad=14)
plt.legend(['Uptown (70115)','Lower Garden District (70130)','Audubon (70118)','Metairie (70005)',
    'Aurora (70131)'])
plt.savefig(r'../images/zip_prices_overview.png')
```
